```python
from sklearn import datasets
iris = datasets.load_iris()
print(iris.data.shape)
print(iris.target.shape)
testx = iris.data[:,[0,1]]
print(testx.shape)
t0 = testx[iris.target == 0]
print(t0.shape)
t1 = testx[iris.target == 1]
print(t1.shape)
t2 = testx[iris.target == 2]
print(t2.shape)
```

```python
import matplotlib.pyplot as plt
plt.scatter(t0[:,0], t0[:,1], c='r')
plt.scatter(t1[:,0], t1[:,1], c='g')
plt.scatter(t2[:,0], t2[:,1], c='b')
plt.show()
```
